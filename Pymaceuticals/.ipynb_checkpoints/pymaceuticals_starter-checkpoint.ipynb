```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)
```

```python
mouse_df.head()
```

```python
clinical_df.head()
```

```python
# Combine the data into a single dataset
merge_outer_df = pd.merge(mouse_df, clinical_df, on='Mouse ID', how='outer')
# Display the data table for preview
merge_outer_df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
df_drug_mean = merge_outer_df[['Drug','Timepoint', 'Tumor Volume (mm3)']].groupby(['Drug','Timepoint'], as_index=False).mean()

# Convert to DataFrame
df_drug_mean
# Preview DataFrame
df_drug_mean.head()
```

```python
merge_outer_df.head()
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
df_tumor_sem = merge_outer_df[['Drug','Timepoint','Tumor Volume (mm3)']].groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)'].to_frame().reset_index()
# Convert to DataFrame
df_tumor_sem.head()
# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_drug_mean = df_drug_mean.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
pivot_drug_mean.head()
```

```python
pivot_drug_sem = df_tumor_sem.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
pivot_drug_sem.head()
```

```python
### Generate the Plot (with Error Bars)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Capomulin"], pivot_drug_sem["Capomulin"], color="r", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Ceftamin"], pivot_drug_sem["Ceftamin"], color="b", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Infubinol"], pivot_drug_sem["Infubinol"], color="g", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Ketapril"], pivot_drug_sem["Ketapril"], color="y", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Naftisol"], pivot_drug_sem["Naftisol"], color="orange", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Placebo"], pivot_drug_sem["Placebo"], color="brown", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Propriva"], pivot_drug_sem["Propriva"], color="black", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Propriva"], pivot_drug_sem["Propriva"], color="black", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Ramicane"], pivot_drug_sem["Ramicane"], color="grey", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Stelasyn"], pivot_drug_sem["Stelasyn"], color="purple", marker ="o", markersize=3, linewidth=0.5)
plt.errorbar(pivot_drug_mean.index, pivot_drug_mean["Zoniferol"], pivot_drug_sem["Zoniferol"], color="aqua", marker ="o", markersize=3, linewidth=0.5)
plt.title('Tumor response to treatment')
plt.xlabel('Time (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.xlim(0,46)
plt.ylim(34,72)
plt.xticks(np.arange(0, 46, 5))
plt.yticks(np.arange(35,75,5))
plt.grid()
# Save the Figure
plt.savefig("tumor_response_to_treatment.pdf")
plt.show()
#Me falta dar formato y poner las legendas
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
df_metastic_mean = merge_outer_df[['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint'], as_index=False).mean()
# Convert to DataFrame
# Preview DataFrame
df_metastic_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
df_metastic_sem = merge_outer_df[['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint']).sem()['Metastatic Sites'].to_frame().reset_index()
# Convert to DataFrame

# Preview DataFrame
df_metastic_sem
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_metastic_mean = df_metastic_mean.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

# Preview that Reformatting worked
pivot_metastic_mean
```

```python
pivot_metastic_sem = df_metastic_sem.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
pivot_metastic_sem
```
